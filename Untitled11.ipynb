```python
outliers = []
import numpy as np
def detect_outlier(data):
  threshold = 3 #3rd std dev
  mean = np.mean(data)
  std = np.std(data)
  for i in data:
    z_score = (i-mean)/std
    if np.abs(z_score)> threshold:
      outliers.append(i)
  return outliers

   

```

```python
detect_outlier(dataset)
```
